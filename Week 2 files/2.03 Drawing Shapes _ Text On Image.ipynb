````python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

img = cv2.imread('media/M1/party.jpg',1)
````



There are many drawing functions in opencv like drawing rectangles , circles , text etc. All of them have the following parameters in common

- `img` : your Image
- `color` : Color of the shape, for BGR image, pass it as a tuple i.e. (255,0,0) for Grayscale image just pass the scalar value.
- `thickness` : Thickness of the line or circle etc. If **-1** is passed for closed figures like circles, it will fill the        shape. default thickness = 1
- `lineType` : Type of line, popular choice is cv2.LINE_AA gives 

###  <font style="color:rgb(134,19,348)">  Drawing  a line on the Image
</font>

We can draw a line in opencv by using the function ``cv2.line()``
 
[```cv2.line(img, pt1, pt2, color, [thickness])```](https://docs.opencv.org/4.2.0/d4/d9d/structcv_1_1gapi_1_1wip_1_1draw_1_1Line.html#aba7af35a1455413aca6c5e1556de6ae7) 
 
 **Params:**
- `pt1:`	First point of the line.
- `pt2:`	Second point of the line.

````python
imgc = img.copy()
cv2.line(imgc,(100,300),(300,50),(255,255,0),5)
plt.imshow(imgc[...,::-1]);
````

###  <font style="color:rgb(134,19,348)">  Drawing  a Circle on the Image
</font>

We can draw a Circle in opencv by using the function ``cv2.Circle()``
 
 [```cv2.circle(img, Center, radius, color, [thickness])```](https://docs.opencv.org/4.2.0/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670) 
 
 **Params:**
- `center:`	Center of the circle.
- `radius:`	Radius of the circle.

````python
imgc = img.copy()
cv2.circle(imgc,(100,63), 55, (0,100,0), 5)
plt.imshow(imgc[...,::-1]);
````

### <font style="color:rgb(134,19,348)">  Drawing  a rectangle on the Image
</font>

We can draw a rectangle in opencv by using the function ``cv2.rectangle()``
 
 [```cv2.rectangle(img, pt1, pt2, color, [thickness])```](https://docs.opencv.org/4.2.0/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9)
 
 **Params:**
- `pt1:`	Top left corner of the rectangle.
- `pt2:`	Top right corner of the rectangle.

````python
imgc = img.copy()
cv2.rectangle(imgc,(384,20),(510,228),(0,255,0),3)
plt.imshow(imgc[...,::-1]);
````

###  <font style="color:rgb(134,19,348)">  Drawing  a polygon on the Image
</font>

We can draw a polygon in opencv by using the function ```cv2.polylines``` <br>
This is really useful as polygon allow us to represent any custom shape
 
 [```img=cv2.polylines(img, pts, isClosed, color, [thickness])```](https://docs.opencv.org/4.2.0/d6/d6e/group__imgproc__draw.html#gaa3c25f9fb764b6bef791bf034f6e26f5)
 
 **Params:**
- `pts:` Array of polygonal curves. Basically its a numpy array of all the points you wish to join.
- `isClosed:` Flag which tells to close the polygon or not. If they are closed, the function draws a line from the last vertex to its first vertex.

````python
imgc = img.copy()
imgc2 = img.copy()

pts = np.array([[10,5],[100,200],[70,340],[300,300]]) #lets make a polygon of 4 points

cv2.polylines(imgc, [pts], True, (255,0,255), 5)   # close this polygon
cv2.polylines(imgc2, [pts], False, (255,0,255), 5)  # left this polygon open

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(imgc[:,:,::-1]);plt.title("Closed Polygon");
plt.subplot(122);plt.imshow(imgc2[:,:,::-1]);plt.title("Open Polygon");
````

### <font style="color:rgb(134,19,348)">  Writing text on the Image
</font>

We can Write Text in opencv by using the function ``cv2.putText`` <br>
Writing Text on images will is an essential tool, you will be able to see real stats on image instead of printing.
 
[```cv.putText(img,text,org,fontFace,fontScale,color,[thickness])```](https://docs.opencv.org/4.2.0/d6/d6e/group__imgproc__draw.html#ga5126f47f883d730f633d74f07456c576)
 
 **Params:**

- `text`	  Text string to be drawn.
- `org`	      Top-left corner of the text (x,y) origin position.
- `fontFace`  Font type, we will use `cv2.FONT_HERSHEY_SIMPLEX`.
- `fontScale` Font scale, how large your text will be.


````python
imgc = img.copy()
cv2.putText(imgc,'Bleed-AI',(0,130),cv2.FONT_HERSHEY_SIMPLEX, 2.7, (0,0,155), 4, cv2.LINE_AA)
plt.imshow(imgc[...,::-1]);
````
