```python
path = "D:/Personal/Images/Umar.png"
img = cv2.imread(path,1)
print (img)
```

```python
print(img.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# You can finally show the image with plt.imshow()
plt.imshow(img)
```

```python
plt.imshow(img,cmap='gray')
```

```python
bgr = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(bgr)
```

```python
#Transforming this to gray scale 
grayScale = cv2.cvtColor(bgr,cv2.COLOR_RGB2BGR)
plt.imshow(grayScale,cmap='gray')
```

```python
plt.imshow(grayScale[:,:,::-1])
```

```python
plt.xticks([])
plt.yticks([])
#plt.imshow(grayScale[:,:,::-1])
plt.imshow(grayScale)
```

```python
plt.figure(figsize=[10,10])
plt.xticks([])
plt.yticks([])
plt.subplot(121)
plt.imshow(bgr)
plt.title("RGB")
plt.subplot(122)
plt.imshow(grayScale)
plt.title("GrayScale")
```

<h1>Drawing on the Images</h1>

```python
cv2.namedWindow("MYImage",cv2.WINDOW_NORMAL)
cv2.imshow("MYImage",grayScale)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow("MYImage",cv2.WINDOW_NORMAL)
img = grayScale.copy()
cv2.line(img,(100,500),(400,0),(255,0,0),5)
cv2.imshow("MYImage",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
%matplotlib notebook
plt.imshow(img)
```

<h1>Drawing Circle

```python
img = grayScale.copy()
cv2.circle(img,(100,63), 55, (0,100,0), 5)
%matplotlib notebook
plt.imshow(img[...,::-1]);
```

```python
img = grayScale.copy()
cv2.circle(img,(295,100), 55, (0,100,0), 5)
%matplotlib notebook
plt.imshow(img[...,::-1]);
```

```python
img = grayScale.copy()
cv2.rectangle(img,(80,10),(550,350),(0,255,0),3)
cv2.imshow('Window',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img  = grayScale.copy()
pts = np.array([[10,5],[100,200],[70,340],[300,300]])
cv2.polylines(img,[pts],True,(255,0,255), 5)
cv2.imshow('window',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img  = grayScale.copy()
cv2.putText(img,"UMAR",(100,300),cv2.FONT_HERSHEY_COMPLEX,1, (0,0,13),1, cv2.LINE_AA)
cv2.imshow('window',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = grayScale.copy()
cv2.rectangle(img,(80,10),(550,350),(0,255,0),3)
cv2.circle(img,(230,150), 80, (0,100,0), 5)

cv2.putText(img,"UMAR",(200,160),cv2.FONT_HERSHEY_COMPLEX,1, (0,0,13),1, cv2.LINE_AA)
cv2.imshow('window',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
